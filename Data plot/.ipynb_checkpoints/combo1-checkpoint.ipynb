```python
from itertools import islice

def window(seq, n=2):
    "Returns a sliding window (of width n) over data from the iterable"
    "   s -> (s0,s1,...s[n-1]), (s1,s2,...,sn), ...                   "
    it = iter(seq)
    result = tuple(islice(it, n))
    if len(result) == n:
        yield result    
    for elem in it:
        result = result[1:] + (elem,)
        yield result

c1_matrix = []     # active power window-sliced matrix
c1q_matrix = []    # reactive power window-sliced matrix

for ts_i in window(ts_c1.P, 4):
    c1_matrix.append(ts_i)
c1_matrix = np.array(c1_matrix)

for ts_i in window(ts_c1q.Q, 4):
    c1q_matrix.append(ts_i)

c1q_matrix = np.array(c1q_matrix)

c1q_matrix.shape
```

```python
# how to describe YD and status in the DataFrame?
#YD_Status
# OFF: -1, YD0: 0.0
# YD1: 1.1, 1.2, 1.3/  YD2: 2.1, 2.2, 2.3, 2.4, 2.5/  YD3: 3.0/  YD4: 4.0, 4.1, 4.2
# YD5: 5.0/  YD6: 6.0/  YD7: 7.0, 7.1, 7.2, 7.3/  YD8: 8.1, 8.2, 8.3/  YD9: 9.1, 9.2, 9.3
# YD10: 10.1, 10.2, 10.3, 10.4/  YD11: 11.0
#Action:
# 1.1, 1.2, 1.3, 


# Save both P and Q matrix into DF
c1_df = pd.concat([pd.DataFrame(data=c1_matrix), pd.DataFrame(data=c1q_matrix)], axis=1)
c1_df.columns = ['P0', 'P1', 'P2', 'P3', 'Q0', 'Q1', 'Q2', 'Q3']
c1_df['Stable'] = 1

z=[]
for i in xrange(len(c1_df)):
    z.append([-1]) 

c1_df['YD'] = z

# use c1_df.iloc[winId]['YD'].append(some label) and c1_df.iloc[winId]['YD'].remove(some label) to assign labels

c1_df.iloc[380:390]
```
